```python
market_data = pd.read_csv("data/market_fact.csv")
customer_data = pd.read_csv("data/cust_dimen.csv")
product_data = pd.read_csv("data/prod_dimen.csv")
```

```python
market_data
```

```python
market_data.head()
```

```python
customer_data.head()
```

```python

product_data.head()
```

```python
new_df = pd.merge(market_data,customer_data,how="inner",on="Cust_id")

## if column names differ then use left_on and right_on
```

```python
new_df
```

```python
final_data = pd.merge(new_df,product_data,how="inner",on="Prod_id")
```

```python
final_data
```

```python
final_data.to_csv("data/Final_Merged_Data.csv",index=False) ### Writing the data to a csv file and save it in system
```
